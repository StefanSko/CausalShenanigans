# Test for Blog Post with Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 3000

u = np.random.normal(0, 1, n_samples)
v = np.random.normal(0, 1, n_samples)
```

```python
plt.scatter(v, u, alpha=0.5)
plt.xlabel('Feature V')
plt.ylabel('Feature U')
plt.title('Test Plot')
```
